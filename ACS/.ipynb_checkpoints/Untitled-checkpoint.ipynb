```python
import pandas as pd 
path = "\\PA"
dps = {}
for i in range(2,6):
    dp_temp = {}
    for x in range(1,18):
        filename = path+ "\\DP%02d_PA%02d.xlsx" %(i,x)
        dp_temp["PA%02d" %x] = pd.read_excel(filename)
        dps["DP%02d" %i] = dp_temp
    print("Imported DP0"+str(i))
```

```python
def getDataVector(table, row , col):
    row = row-2
#     print(dps[table]["PA01"]["Table ID: "+table][row])
    rtn = []
    for x in range(1,18):
        df = dps[table]["PA%02d" %x]
#         print("PA%02d" % x ,"\t",df["Unnamed: "+str(1)][row])  
        rtn.append(df["Unnamed: "+str(col)][row])
    return dps[table]["PA01"]["Table ID: "+table][row] , rtn
```

```python
vectors = []
things = [ ("DP05" , 52 , 3) , ("DP02" , 93 , 3) ,("DP03",84,1) , ("DP03",53,3) ,("DP03",60,3) ,
#             White                  HS             median income  Manufacturing   Educaiton      
          ("DP03" , 83 , 3),
                          ("DP03" , 85 , 1) ,
                                                  ("DP03" , 128 , 3),("DP02" , 130 , 3) ,("DP02" , 159 , 3),
#           200+              Meanincome             public health       foreign born      english only           
         ("DP02" , 204 , 3) ,
                                  ("DP03" , 43 , 3),
                                                       ("DP03" , 37 , 3)]
#               internet           job                 work from home
for table , row , col in things:
    vectors.append(getDataVector(table,row,col))


df = pd.DataFrame({
    "District" : ["PA%02d" %x for x in range(1,18)],
    **{vectors[x][0] : vectors[x][1] for x in range(0,len(vectors))}
})


# df = df.drop('$200,000 or more', axis=1)
# df = df.drop('Mean household income (dollars)', axis=1)
# df = df.drop('Management, business, science, and arts occupations', axis=1)


df
```

```python
import numpy as np
from sklearn.preprocessing import minmax_scale
from sklearn.metrics.pairwise import cosine_similarity
vector_scaled = np.array([minmax_scale(x[1]) for x in  vectors])

edge_list = []
threshold = .9
skip_list = []
for i in range(1,18):
    skiped = True
    v1 = vector_scaled.T[i-1]
    for j in range(i+1,18):
        v2 = vector_scaled.T[j-1]
        score =  cosine_similarity(v1.reshape(1, -1) , v2.reshape(1, -1))
        print(i,j,score)
        if score > threshold:
            edge_list.append(("PA"+str(i) , "PA"+str(j)))
            skiped = False
    if skiped :
        skip_list.append ("PA"+str(i))

```
